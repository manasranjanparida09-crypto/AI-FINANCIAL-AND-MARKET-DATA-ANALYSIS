### How much amount the companies spent on R & D ?

```python
print("Company's spending for R & D in $Bn")
R_D=df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000   # Billion Dollar
R_D
```

```python
#Draw a Bar plot to show the amount spent on R & D by the Companies
plt.bar(R_D.index,R_D.values,color=['cyan','black','magenta'])

plt.title('R&D Spending by Companies')
plt.xlabel('Company')
plt.ylabel("Amount in USD($Bn)")
plt.show()
```

```python
##Revenue Earned By the Companies
df.head(3)
```

```python
print("Company's AI Revenue (USD_$Bn)")
air=df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000 ##Billion dollar
air
```

```python
#Draw a Bar Plot to show the Revenues of the companies
plt.bar(air.index,air.values,color=['cyan','black','magenta'],width=0.8)

plt.title('AI revenue by Companies')
plt.xlabel('Company')
plt.ylabel("AI_Revenue  in USD($Bn)")
plt.show()
```

```python
#BAr plot to show expenditure & Revenue of the companies
#Subplot function
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)

plt.bar(R_D.index,R_D.values,color=['cyan','black','magenta'])

plt.title('R&D Spending by Companies')
plt.xlabel('Company')
plt.ylabel("Amount in USD($Bn)")

plt.subplot(1,2,2)

plt.bar(air.index,air.values,color=['cyan','black','magenta'],width=0.8)

plt.title('AI revenue by Companies')
plt.xlabel('Company')
plt.ylabel("AI_Revenue  in USD($Bn)")
plt.show()
```

### Expenditure vs Revenue year-by-year

```python
df.head(5)
```

```python
spend=df.groupby('Year')['R&D_Spending_USD_Mn'].sum()
spend
```

```python
plt.plot(spend.index,spend.values,color='r')

plt.title('Combined R&D Spending Year by Year ')
plt.xlabel('Year')
plt.ylabel("Amount in USD_$Mn")
plt.show()
```

```python
revenue=df.groupby('Year')['AI_Revenue_USD_Mn'].sum()
revenue
```

```python
plt.plot(revenue.index,revenue.values,color='b')

plt.title('Combined Revenue Earned Year by Year ')
plt.xlabel('Year')
plt.ylabel("Amount in USD_$Mn")
plt.show()
```

```python
plt.plot(spend.index,spend.values,color='r')
plt.plot(revenue.index,revenue.values,color='b')

plt.title('Combined Expenditure vs Revenue Year by Year ')
plt.xlabel('Year')
plt.ylabel("Amount in USD_$Mn")
plt.legend(['Expenditure','Revenue'])
plt.show()
plt.show()
```
